```python
#matplotlib.use('TkAgg')
xs = torch.linspace(-1, 1, steps=300)
ys = torch.linspace(-1, 1, steps=300)
x, y = torch.meshgrid(xs, ys, indexing="xy")
```

```python
x = np.linspace(-1,1,300)
xs = np.array(np.meshgrid(x,x))
points = np.transpose(xs.reshape(2,-1))

p_tensor = torch.from_numpy(points)
#print(p_tensor)
affine = torch.tensor([[2, 2], [2, 0], [0, 2]], dtype=torch.float64)
result = torch.zeros((300*300,3))
for i, val in enumerate(p_tensor):
    #print(val)
    result[i] = torch.matmul(affine, val)
    #print(torch.matmul(affine, val))
z = torch.max(result, dim=1)[0]
#print(z)
z = torch.reshape(z, (300,300))
xs = torch.linspace(-1, 1, steps=300)
ys = torch.linspace(-1, 1, steps=300)
x, y = torch.meshgrid(xs, ys, indexing="xy")
ax = plt.axes(projection="3d")
ax.contourf(x.numpy(), y.numpy(), z.numpy(), 200)
plt.show()
```
